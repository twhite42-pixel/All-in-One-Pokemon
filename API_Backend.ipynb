# PokeAPI Setup

## Package Setup
We need to connect the api and some other packages to view data.  
The requests package allows us to send HTTP web requests and make API calls.

```python
import requests
import json
from IPython.display import Image, Audio, display

# This url is used to request data from PokeAPI. Add more specific endpoints to this url to request more specific data.
base_url = "https://pokeapi.co/api/v2/"
```

## Basic Pokemon Info
This function inputs the pokemon that we want into the api in a usable format and gets some basic information about that pokemon.

```python
# Gets information of a specific Pokemon. pokemon_name is now the value of (name) in the function call.
def get_pokemon_info(name):
    url = f"{base_url}/pokemon/{name}"
    response = requests.get(url)

    # For successful API calls, response code is 200.
    if response.status_code == 200:
        # Save data to a json response and return data as a dictionary.
        pokemon_data = response.json()
    # Save data to a json file to view data structure(Only for exploration)
        # with open(f"{name}.json", "w") as f:
            # json.dump(pokemon_data, f, indent=4)
        return pokemon_data
        
        
    # For unsuccessful API calls, print the error code.    
    else:
        print(f"Failed to retrieve data {response.status_code}")
```

## Dictionary for Pokemon Stats
This function maps stats to particular categories to display in a nicer format

```python
stat_names = {
    "hp": "HP",
    "attack": "Attack",
    "defense": "Defense",
    "special-attack": "Special Attack",
    "special-defense": "Special Defense",
    "speed": "Speed"
}
```

### Example Function Call
This function gets basic information about a certain pokemon. It also returns a dictionary of information.

```python
pokemon_name = input("Enter a Pokemon name: ").lower()
pokemon_info = get_pokemon_info(pokemon_name)
artwork_url = pokemon_info['sprites']['other']["official-artwork"]['front_default']
sprite_url = pokemon_info['sprites']['front_default']
cry_url = pokemon_info['cries']['latest']

if pokemon_info:
    # Show the default sprite of Pokemon and can hear the cry.
    display(Image(url=artwork_url))
    display(Image(url=sprite_url))
    display(Audio(url=cry_url))
    print(f"Name: {pokemon_info['name'].capitalize()}")
    print(f"ID: {pokemon_info['id']}")
    print(f"Height: {pokemon_info['height']}")
    print(f"Weight: {pokemon_info['weight']}")
    # Shows all possible abilities for Pokemon.
    print(f"Abilities: {", ".join([ability_info['ability']['name'].replace('-', ' ').title() 
        for ability_info in pokemon_info['abilities']])}")
    # Shows types for Pokemon. If multiple types, they are separated by commas.
    print(f"Type(s): {', '.join([type_info['type']['name'].capitalize() for type_info in pokemon_info['types']])}")
    # Shows all possible held items for Pokemon. If none, print "None"
    print(f"Held Items: {', '.join([item_info['item']['name'].replace('-', ' ').title()
        for item_info in pokemon_info['held_items']]) if pokemon_info['held_items'] else 'None'}")
    # Show all base stats for Pokemon. Formatted category names nicely.
    print(f"Stats: {', '.join([f'{stat_names[stat_info['stat']['name']]}: {stat_info['base_stat']}' for stat_info in pokemon_info['stats']])}")
    # Shows all possible moves for Pokemon. Sorted in alphabetical order and formatted nicely.
    print(f"Moves: {', '.join(
        sorted(
            move_info['move']['name'].replace('-', ' ').title() 
            for move_info in pokemon_info['moves']
        ))}"
    )
```

## Item Function Call
Similar functionality to provide relevant information about items

```python
# Gets information of a specific item. item_name is now the value of (name) in the function call.
def get_item_info(name):
    url = f"{base_url}/item/{name}"
    item_response = requests.get(url)

    # For successful API calls, response code is 200.
    if item_response.status_code == 200:
        # Save data to a json response and return data as a dictionary.
        item_data = item_response.json()
    # Save data to a json file to view data structure(Only for exploration)
        # with open(f"{name}.json", "w") as f:
            # json.dump(item_data, f, indent=4)
        return item_data


    # For unsuccessful API calls, print the error code.
    else:
        print(f"Failed to retrieve data {item_response.status_code}")
```

```python
item_name = input("Enter an item name: ").replace(' ', '-').lower()
item_info = get_item_info(item_name)
item_sprite_url = item_info['sprites']['default']

if item_info:
    # Show the default sprite of the item.
    display(Image(url=item_sprite_url))
    print(f"Name: {item_info['name'].replace('-', ' ').title()}")
    print(f"ID: {item_info['id']}")
    print(f"Cost: {item_info['cost']}")
    print(f"Category: {item_info['category']['name'].replace('-', ' ').title()}")
    # Shows short and full effect of the item. [1] is used to get the English entry.
    print(f"Short Effect: {item_info['effect_entries'][1]['short_effect']}")
    print(f"Full Effect: {item_info['effect_entries'][1]['effect']}")
```

## Sprite Calls
Simple functions to call different kinds of images. This can include various sprites, official artwork, cries, and other calls which only require an input of a pokemon name.

```python
def get_pokemon_sprite(pokemon_name):
    url = f"{base_url}/pokemon/{pokemon_name}"
    response = requests.get(url)

    if response.status_code == 200:
        pokemon_data = response.json()
        sprite_url = pokemon_data['sprites']['front_default']
        display(Image(url=sprite_url))
        return sprite_url
    
    else:
        print(f"Failed to retrieve data {response.status_code}")
        return None
    
def get_pokemon_artwork(pokemon_name):
    url = f"{base_url}/pokemon/{pokemon_name}"
    response = requests.get(url)

    if response.status_code == 200:
        pokemon_data = response.json()
        artwork_url = pokemon_data['sprites']['other']["official-artwork"]['front_default']
        display(Image(url=artwork_url))
        return artwork_url
    
    else:
        print(f"Failed to retrieve data {response.status_code}")
        return None
    
def get_home_artwork(pokemon_name):
    url = f"{base_url}/pokemon/{pokemon_name}"
    response = requests.get(url)

    if response.status_code == 200:
        pokemon_data = response.json()
        home_artwork_url = pokemon_data['sprites']['other']["home"]['front_default']
        display(Image(url=home_artwork_url))
        return home_artwork_url
    
    else:
        print(f"Failed to retrieve data {response.status_code}")
        return None
    
def get_pokemon_cry(pokemon_name):
    url = f"{base_url}/pokemon/{pokemon_name}"
    response = requests.get(url)

    if response.status_code == 200:
        pokemon_data = response.json()
        cry_url = pokemon_data['cries']['latest']
        display(Audio(url=cry_url))
        return cry_url
    
    else:
        print(f"Failed to retrieve data {response.status_code}")
        return None

def get_item_sprite(item_name):
    url = f"{base_url}/item/{item_name}"
    response = requests.get(url)

    if response.status_code == 200:
        item_data = response.json()
        sprite_url = item_data['sprites']['default']
        display(Image(url=sprite_url))
        return sprite_url
    
    else:
        print(f"Failed to retrieve data {response.status_code}")
        return None
```
