# PokeAPI Setup

## Package Setup
We need to connect the api and some other packages to view data.  
The requests package allows us to send HTTP web requests and make API calls.

```python
import requests

# This url is used to request data from PokeAPI. Add more specific endpoints to this url to request more specific data.
base_url = "https://pokeapi.co/api/v2/"
```

## Basic Pokemon Info
This function inputs the pokemon that we want into the api in a usable format and gets some basic information about that pokemon.

```python
# Gets information of a specific Pokemon. pokemon_name is now the value of (name) in the function call.
def get_pokemon_info(name):
    url = f"{base_url}/pokemon/{name}"
    response = requests.get(url)

    # For successful API calls, response code is 200.
    if response.status_code == 200:
        # Save data to a json response and return data as a dictionary.
        pokemon_data = response.json()
        return pokemon_data
        
        
    # For unsuccessful API calls, print the error code.    
    else:
        print(f"Failed to retrieve data {response.status_code}")
```

### Example Function Call
This function gets basic information about a certain pokemon. It also returns a dictionary of information.

```python
pokemon_name = "ditto"
pokemon_info = get_pokemon_info(pokemon_name)

if pokemon_info:
    print(f"Name: {pokemon_info['name'].capitalize()}")
    print(f"ID: {pokemon_info['id']}")
    print(f"Height: {pokemon_info['height']}")
    print(f"Weight: {pokemon_info['weight']}")
```
